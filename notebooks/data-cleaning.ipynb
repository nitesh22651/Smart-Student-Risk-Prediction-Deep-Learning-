## Step 4: Data Cleaning

```python
# Drop unwanted columns
cols_2_drop = [
    'Student_ID', 'First_Name', 'Last_Name', 'Email', 'Age', 'Gender', 
    'Department', 'Midterm_Score', 'Final_Score', 'Participation_Score', 
    'Projects_Score', 'Grade', 'Sleep_Hours_per_Night', 'Internet_Access_at_Home', 
    'Parent_Education_Level', 'Family_Income_Level'
]
df.drop(columns=cols_2_drop, inplace=True)
```

```python
df.sample(3)
```

```python
# Rename columns
new_cols = {
    "Attendance (%)": "attendance_percentage",
    "Assignments_Avg": "assignments_rate",
    "Quizzes_Avg": "quizzes_score",
    "Total_Score": "previous_grades",
    "Study_Hours_per_Week": "time_spent_on_study",
    "Extracurricular_Activities": "extracurricular",
    "Stress_Level (1-10)": "stress_level"
}
df.rename(columns=new_cols, inplace=True)
df.sample(3)
```

## Step 5: Create the new column 'dropout_confidense'

```python
for c in df.columns:
    if df[c].dtype.kind in 'biufc':  # numeric
        df[c] = df[c].fillna(df[c].median())
    else:
        df[c] = df[c].astype('category').cat.codes
        df[c] = df[c].fillna(df[c].median())

# Normalize each feature to 0-1
def minmax(series):
    if series.max() == series.min():
        return pd.Series(0.5, index=series.index)  # constant column -> neutral 0.5
    return (series - series.min()) / (series.max() - series.min())

att = minmax(df['attendance_percentage'])
assn = minmax(df['assignments_rate'])
quiz = minmax(df['quizzes_score'])
prev = minmax(df['previous_grades'])
time_spent = minmax(df['time_spent_on_study'])
extra = minmax(df['extracurricular'])
stress = minmax(df['stress_level'])
```

```python
# Weights (sum to 1)
w = {
    'attendance': 0.25,
    'assignments': 0.20,
    'quizzes': 0.15,
    'previous': 0.15,
    'time_spent': 0.15,
    'extracurricular': 0.05,
    'stress': 0.05
}

# Compute risk: higher when attendance, assignments, quizzes, previous grades, time_spent, extracurricular are LOW
# and when stress is HIGH.
np.random.seed(42)
base_risk = (
    w['attendance']*(1 - att) +
    w['assignments']*(1 - assn) +
    w['quizzes']*(1 - quiz) +
    w['previous']*(1 - prev) +
    w['time_spent']*(1 - time_spent) +
    w['extracurricular']*(1 - extra) +
    w['stress']*(stress)  # stress increases risk
)

# Add a little non-deterministic noise and clip to [0,1]
noise = np.random.normal(loc=0.0, scale=0.05, size=len(df))
dropout_confidense = (base_risk + noise).clip(0,1)

# Attach to dataframe
df['dropout_confidense'] = dropout_confidense.round(4)
```
