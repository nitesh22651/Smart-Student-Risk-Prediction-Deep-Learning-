```python
import joblib
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import load_model
```

## Step 2: Load Model

```python
preprocessor = joblib.load("../artifacts/preprocessor.joblib")
model = load_model("../artifacts/checkpoints/resnet_tabular_best.keras")  
```

```python
numeric_features = [
    "attendance_percentage",
    "assignments_rate",
    "quizzes_score",
    "previous_grades",
    "time_spent_on_study",
    "extracurricular",
    "stress_level",
]

def preprocess_input(data: dict) -> np.ndarray:
    """
    Convert input dict to a row for the sklearn preprocessor.
    """
    row_df = pd.DataFrame(
        [{feat: float(data[feat]) for feat in numeric_features}],
        columns=numeric_features,
    )
    X_processed = preprocessor.transform(row_df)  # returns numpy array
    return np.asarray(X_processed, dtype=np.float32)


def predict_dropout(data: dict) -> float:
    """
    Predict dropout confidence using sklearn preprocessor + .keras model.
    """
    x = preprocess_input(data)
    y_pred = model.predict(x, verbose=0)
    return float(y_pred[0][0])
```

```python
prediction = predict_dropout({
    "attendance_percentage": 87.0,
    "assignments_rate": 50.0,
    "quizzes_score": 75,
    "previous_grades": 60,
    "time_spent_on_study": 20,
    "extracurricular": 1,
    "stress_level": 5,
})

print(f"Course Dropout Confidense: {prediction:.2f}")
```
